# **DATA PREPROCESSING**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
mushroom['class'] = le.fit_transform(mushroom['class'])
```

# **SVM implementation**

```python
from sklearn.svm import SVC
svc=SVC(C=5,gamma=4,kernel='poly')
```

```python
from sklearn.model_selection import train_test_split

X_train ,X_test ,Y_train , Y_test = train_test_split(X,Y,test_size=0.25 , random_state = 10)
```

```python
svc.fit(X_train,Y_train)
```

```python
ypred = svc.predict(X_test)
ypred
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,ypred))
```

```python
print(svc.score(X_train,Y_train),svc.score(X_test,Y_test))
```

**2**. ***Strengths and Weaknesses***


**Strengths:**

* SVM is effective in handling high-dimensional data with a large number of features.
* It can handle non-linear relationships between features and the target variable.
* SVM is robust to outliers and noise in the data.
* It provides a unique solution due to its convex optimization problem formulation.

**Weaknesses:**

* SVM can be computationally expensive for large datasets.
* The choice of the kernel function and its parameters can significantly affect the performance of the model.
* SVM is sensitive to the scale of the features.
* It may not perform well with overlapping classes.

**Specific to the Mushroom dataset:**

* SVM performed well on the Mushroom dataset, achieving an accuracy of 59.02% after hyperparameter tuning.
* The confusion matrix showed that the model had difficulty distinguishing between edible and poisonous mushrooms.
* The ROC curve indicated that the model had a good ability to distinguish between the two classes.
* Overall, SVM was a suitable choice for the Mushroom dataset, given its ability to handle high-dimensional data and non-linear relationships.

**Further analysis:**

* Additional techniques such as feature engineering and dimensionality reduction could be explored to improve the performance of the model.
* Different kernel functions and parameters could be tested to find the optimal configuration for the Mushroom dataset.
* Other classification algorithms could be compared with SVM to determine the best approach for this specific problem.
